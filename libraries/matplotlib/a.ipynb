```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x= [1,2,3,4,5]
y= [4,6,8,6,3]
y2= [9,6,8,7,4]
plt.plot(x,y, color="red", linestyle = "--", marker= ">", markersize=25)
plt.plot(x,y2, color="blue", linestyle = "--" , marker= ">", markersize=25)
plt.title("line chart")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
#bar chart
x=["A", "B", "c", "D"]
y=[10, 15, 7, 12]
plt.bar(x,y, color="gold")
plt.show()
```

```python
#horizontal bar chart
x=["A", "B", "c", "D"]
y=[10, 15, 7, 12]
plt.barh(x,y, color="orange")
plt.show()
```

```python
x=["python", "java", "c++", "js"]
y=[40, 25,20, 15]
plt.pie(y, labels=x, autopct='%1.1f%%', startangle=180 )
plt.show()
```

```python
data= np.random.randn(100)
plt.hist(data, bins=30, color='orange')
plt.show()
```

```python
x= [1,2,3,4,5]
y=[5,7,9,3,5]
plt.scatter(x, y, color='r', s=100, marker="*")
plt.savefig("sactter.png")
plt.show()
```
